# The Erdős Unit-Distance Problem

## Definition

Let u(n) be the largest number of unordered pairs at Euclidean distance exactly 1 among n planar points.

## Key Sources

1. **OpenAI (2026-05-20):** AI-generated counterexample
2. **Alon et al. (arXiv:2605.20695):** Human-verified companion paper
3. **Sawin (arXiv:2605.20579):** Explicit n^{1.014} bound

## What Remains Open

The exact asymptotic determination of u(n) is still open. The gap between n^{1.014} and O(n^{4/3}) is large.
