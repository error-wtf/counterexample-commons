```python
import sys, os
from pathlib import Path

# Resolve repo root whether running in Colab, locally, or via nbconvert
_nb_dir = Path(globals().get("__vsc_ipynb_file__", __file__) if "__file__" in dir() else ".").resolve()
_repo_root = next((p for p in [_nb_dir, *_nb_dir.parents] if (p / "counterexample_commons").is_dir()), None)
if _repo_root and str(_repo_root) not in sys.path:
    sys.path.insert(0, str(_repo_root))

os.environ.setdefault("COUNTEREXAMPLE_MODE", "colab-public-demo")
```

```python
from app.main import build_app
from counterexample_commons.config import AppMode

mode_str = os.environ.get("COUNTEREXAMPLE_MODE", "colab-public-demo")
mode = AppMode(mode_str)
demo = build_app(mode=mode)

IS_COLAB = "google.colab" in sys.modules
if IS_COLAB:
    demo.launch(share=True)
else:
    print(f"App built successfully in mode: {mode.value}")
    print("Run demo.launch() interactively to start the server.")
```
